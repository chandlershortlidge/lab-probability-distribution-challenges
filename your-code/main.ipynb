# Challenge 1 - Plot Probability Mass Function (PMF)

#### Background knowledge

[PMF](https://en.wikipedia.org/wiki/Probability_mass_function) shows the probability distribution of a **discrete random variable**. A [discrete random variable](https://en.wikipedia.org/wiki/Random_variable#Discrete_random_variable) has random numeric values that are not continuous. For example, the number of people in a household can only be integers but not floats. Therefore the number of people in a household is a discrete variable.

#### Question: We assume that the probability of clicking an Ad in a Youtube video is 0.15. We have a sample of 5 people who wathched the video and we want to plot the PMF for a binomial distribution.

*Hint: use binom from `scipy.stats.binom`. Your output should look like below:*

![binom 10](ch-2.png)

```python
# Parameters
n = 5   # number of trials (people)
p = 0.15  # probability of success (click)

# Support (possible values of X: 0 to n)
x = list(range(n+1))

# PMF values
pmf = binom.pmf(x, n, p)

# Plot
plt.bar(x, pmf, color="skyblue", edgecolor="black")
plt.xlabel("Number of Clicks")
plt.ylabel("Probability P(X=k)")
plt.title("Binomial Distribution PMF (n=5, p=0.15)")
plt.show()
```

#### Explain what you observe from the plot above

```python
# There’s about a 45% chance nobody clicks. 
# There’s about a 39% chance exactly one person clicks.
# Roughly 14% chance exactly two people click.
# Higher values (3, 4, 5 clicks) are increasingly rare.
```

#### Now plot PMP with 50, 500, and 5000 visitors.

```python
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
p = 0.15
n_values = [50, 500, 5000]

plt.figure(figsize=(12, 8))

for i, n in enumerate(n_values, 1):
    x = list(range(n+1))
    pmf = binom.pmf(x, n, p)
    
    plt.subplot(3, 1, i)
    plt.plot(x, pmf, color="blue")
    plt.xlabel("Number of Clicks")
    plt.ylabel("P(X=k)")
    plt.title(f"Binomial PMF (n={n}, p={p})")

plt.tight_layout()
plt.show()
```

#### What did you notice from the distribution plots? Comment your findings.

```python
# n = 50: Outcomes are messy, sometimes 5 clicks, sometimes 12.
# n = 500: scaling up to 500 visitors smooths outcomes, usually near 75 clicks.
# With 5000 visitors, 750 clicks are very close to guaranteed — the randomness “averages out.”
```

#### A website has an average of 300 visits per day. What is the probability of getting 320 visitors in a day?

*Hint: use `scipy.stats.poisson`*

```python
from scipy.stats import poisson

# Parameters
lmbda = 300   # average visits
k = 320       # target number of visits

# Poisson probability
p_320 = poisson.pmf(k, lmbda)

print("P(X = 320 visitors) =", p_320)
```

#### What is the probability of getting 60 visits?

```python
from scipy.stats import poisson

lmbda = 300
k = 60

p_60 = poisson.pmf(k, lmbda)
print("P(X = 60 visitors) =", p_60)
```

```python
# Parameters
lmbda = 300
max_visits = 1000

# Values of k
k_values = list(range(max_visits+1))

# Use scipy's poisson.pmf
poisson_probs = poisson.pmf(k_values, lmbda)

# Plot
plt.plot(k_values, poisson_probs, color="blue")
plt.xlabel("Number of Visits (k)")
plt.ylabel("Probability P(X=k)")
plt.title("Poisson Distribution of Daily Visits (λ = 300)")
plt.show()
```
